```python
import requests

def get_usd_to_pkr_rate():
    url = "https://api.exchangerate-api.com/v4/latest/USD"
    response = requests.get(url)
    data = response.json()
    return data["rates"]["PKR"]

rate = get_usd_to_pkr_rate()
amount = float(input("Enter USD amount: "))
print("1 Usd =",rate,"pkr")
print("PKR:", amount * rate)
```
